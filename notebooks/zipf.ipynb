## Zipf's Law

Zipf's distribution is the (approximate) distribution of all words in any language or corpus

Zipf's law is a power law that states the following
> The $r$th most frequent word has a frequency $f(r)$ that scales accordingly to
> $ f(r) \propto \frac{1}{r^\alpha}$ for $\alpha \approx 1$.

Here, 
- $r$ is the **frequency rank** of a word
- $f(r)$ is the words frequency in a natural corpus

I.e. frequency of an item or event is inversely proportional to its frequency rank.

### Log-log scale

Log-log uses logarithmic scales on both x and y-axes.

- *Monomials* i.e. relationships of the form $y = ax^k$ appear as straight lines in a log-log graph
    - $k$ corresponds to the **slope** of the line
    - constant $a$ corresponds to the **intercept** of the line
- Log-log graphs are useful for recognizing these relationships and estimating parameters

Note: For Zipf's distribution $a = 1, x = r$ and $k = -1$


 

### Fitting Zipf in Python

```python
data = [(60462, 'Italy'), (46755, 'Spain'), (10423, 'Greece'), (10197, 'Portugal'), (8737, 'Serbia'), (4105, 'Croatia'),
       (3281, 'Bosnia and\nHerzegovina'), (2878, 'Albania'), (2083, 'North\nMacedonia'), (2079, 'Slovenia'),
       (628, 'Montenegro'), (442, 'Malta'), (77, 'Andorra'), (34, 'San Marino'), (34, 'Gibraltar'), (1, 'Holy See')]

freq = []
countries = []
for f, c in data:
    freq.append(f)
    countries.append(c)
rank = np.array(range(len(freq))) + 1

alpha = 1.37065874
total = sum(freq)
plt.plot(rank-1, sts.zipf.pmf(rank, alpha) * total, lw=1, color='red')
plt.bar(countries, freq)
plt.xticks(rotation='vertical')
plt.show()
```
